```python
y_data=df['price']
```

```python
x_data=df.drop('price',axis=1)
```

```python
x_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.10, random_state=1)

print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])

```

```python
lre=LinearRegression() #Creating a LR object 
```

```python
lre.fit(x_train[['horsepower']], y_train) #Fit 
```

```python
lre.score(x_test[['horsepower']], y_test) #R2 testing - test data
```

```python
lre.score(x_train[['horsepower']], y_train) #R2 Testing - training data
```

```python
Rcross = cross_val_score(lre, x_data[['horsepower']], y_data, cv=4)
Rcross

#output is 'The Average R2 on the test data for each of the two folds'
```

```python
#Mean and Standard Deviation

print("The mean of the folds are", Rcross.mean(), "and the standard devation is", Rcross.std())
```
